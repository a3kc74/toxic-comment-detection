```python
from datasets import load_dataset
import pyarrow.parquet as pq
import pandas as pd
import tools.preprocessing as preprocess
from pathlib import Path
```

```python
ds = load_dataset("nqdhocai/toxic-comment-detection")
print(ds.keys())
```

```python
# -----------------------------
# 3. Áp dụng lên DataFrame
# -----------------------------

df = ds["train"].to_pandas()
# Có 2 dòng comment_text bị NaN -> xử lý:
# Xoá các dòng có comment_text bị NaN
df = df.dropna(subset=["comment_text"])   
df = df.reset_index(drop=True)

# Tạo cột mới cho text đã preprocessing
df["comment_text"] = df["comment_text"].apply(preprocess.preprocess_comment)
# Lưu lại nếu cần
folder = Path("data")
df.to_parquet(folder / f"cleaned_jigsaw.parquet", index=False)
```
